```python
import math
from sklearn.model_selection import train_test_split

def most_common(lst):
    return  max(set(lst), key=lst.count)

def euc_dist(x1,y1,x2,y2,z1,z2):
    ans=math.sqrt((x1-x2)**2 + (y1-y2)**2+(z1-z2)**2)
    return ans

def knn(X_train,xt,yt,zt,y):
    nei=[]
    for i in range(len(X_train)) :
        dist=euc_dist(X_train.iloc[i][2],X_train.iloc[i][1],xt,yt,X_train.iloc[i][0],zt)
        nei.append([dist,y.iloc[i][0]])
    y_sorted=[y for _,y in sorted(nei)]    
    # print(nei)
    return most_common(y_sorted[:5])
           
# df = pd.read_csv('fruit_dataset.csv')
x=df[['mass','width','height']]
y=df[['fruit_label']]
# print(y)
plt.scatter(df['height'],df['width'],c=df['fruit_label'])
# plt.show()
# display(X_test.iloc[0])
X_train,X_test,Y_train, y_test = train_test_split(x,y,random_state=1000,test_size=0.20,shuffle=True)
y_predict=[knn(X_train,X_test.iloc[i][2],X_test.iloc[i][1],X_test.iloc[i][0],y) for i in range(len(X_test))]
count=0
for i in range(len(y_predict)):
    print(y_predict[i], y_test.iloc[i][0])
    if y_predict[i] == y_test.iloc[i][0]:
        count += 1
accuracy = count / len(y_predict) * 100
  
print("Acc = ", accuracy)
# print(y_test)
# print("label",y_test.iloc[0])
```
